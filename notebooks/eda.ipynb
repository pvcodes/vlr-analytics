```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("VCT EDA").master("local[*]").getOrCreate()
```

```python
base_path = "/Users/pvcodes/Data/Codes/dev/vct-analytics/data/bronze"
df = (
    spark.read.option("header", "true")
    .option("inferSchema", "true")
    .option("basePath", base_path)
    .csv(base_path)
)
```
